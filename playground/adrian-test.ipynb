```python
def __get_earliest_timeslot(self, available_time_slots_df:pd.DataFrame) -> pd.DataFrame:
    """
    Looks at the available_time_slots dataframe and retrieves information about the
    earliest available timeslot

    Args:
        available_time_slots_df (pd.DataFrame): contains list of available timeslots for appointments

    Returns:
        pd.DataFrame: information about the earliest appointment
    """
    available_time_slots_df['START_DATETIME'] = pd.to_datetime(available_time_slots_df['START_DATETIME'])

    earliest_appointment = available_time_slots_df.loc[available_time_slots_df['START_DATETIME'].idxmin()]

    return earliest_appointment
```
